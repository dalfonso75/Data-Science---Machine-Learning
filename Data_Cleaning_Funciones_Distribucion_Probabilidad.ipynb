## Distribución Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
a=1 #limite inferior.
b=100 #limite Superior.
n=1000000 # #de Muestras.
data = np.random.uniform(a, b, n)
```

```python
plt.hist(data)
```
